```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#read in the data
df=pd.read_csv('movies.csv')
```

```python
#let's Look at the data
df.head()
```

```python
#let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
df.isnull().sum()
```

```python
#filling the null values
df2=df.fillna(value=0)
```

```python
df2.isnull().sum()
```

```python
#data types for our columns

df2.dtypes
```

```python
# change data typr of columns

df2['budget'] = df2['budget'].astype('int64')
df2['gross'] = df2['gross'].astype('int64')
```

```python
df2
```

```python
df2.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df2['budget'],y=df2['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')

plt.show()
```

```python
correlation_matrix = df2.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

#very dark color is low correlation and bright colors are high correlation in this heatmap
```

```python
df2_numerized = df2

for col_name in df2_numerized.columns:
    if(df2_numerized[col_name].dtype == 'object'):
        df2_numerized[col_name]=df2_numerized[col_name].astype('category')
        df2_numerized[col_name]=df2_numerized[col_name].cat.codes
df2_numerized
```
